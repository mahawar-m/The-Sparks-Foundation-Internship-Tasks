### **States with highest number of buyers**


```python
plt.figure(figsize=(20,15))
sns.countplot(x='State',data=df,palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

####  California and New York have the highest numbers of buyers.

*****

### Analysing Sales, Quantity Purchased, Discount, and Profit

### **Profit vs Discount**

```python
sns.scatterplot(x = 'Profit', y='Discount', data = df)
```

```python
sns.lineplot(y = 'Profit', x='Discount', data = df)
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x='Sales', y='Profit', data = df, hue = 'Discount')
```

#### There is no direct relation between discount and profits. Initially as the discount increases upto 20 percent the profits are positive and it increases. But after the discount of more than 20 percent, the profits decline.

****

### **Segment Profit & Sales**

```python
plt.figure(figsize=(15,10))
df.groupby('Segment')[['Profit', 'Sales']].sum().plot.bar(color=['Red','Orange'], figsize=(15,10))
plt.ylabel('Profit / Sales')
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x='Quantity', y='Profit', data = df)
```

#### The profits and sales are maximum in Consumer segment and minimum in Home Office segment.

****

### Segement wise sales analysis

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Region', y='Sales', data = df, hue ='Segment', palette='rocket' )
plt.title('Segment wise sales in each region')
```

#### The segment wise sales in each region is somewhat similar with nothing important change.

****

### **Sales in each region**

```python
plt.figure(figsize=(15,10))
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(figsize=(15,10))
```

#### The profits and sales are maximum in West region and minimum in South region


****

### Profit & Loss Sales across States

```python
df1 = df.groupby('State')[['Profit','Sales']].sum().sort_values(by='Sales', ascending= False)
df1.plot.bar(color=['Blue', 'Orange'], figsize=(15,10))
plt.ylabel('Profit / Sales')
plt.title('Profit/Loss and Sales across States')
```

#### The Sales and Profits are maximum in the states of California and New York. The States of Texas, Pennsylvania and Ohio on the other hand results in overall loss.

****

### Category Analysis

```python
df.groupby('Category')[['Profit','Sales']].sum().sort_values(by='Sales', ascending= False).plot.bar(color=['Blue','Orange'], figsize=(15,10))
plt.ylabel('Profits / Sales')
```

####  Technology and Office Supplies have high profit. Furniture have less profit

****

### Sub-Category Analysis
